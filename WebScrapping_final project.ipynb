```python
from selenium import webdriver
import time
driver=webdriver.Chrome("/tmp/chromedriver")

driver.get('https://raritysniper.com/nft-collections')


time.sleep(3)
previous_height=driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    time.sleep(3)
    
    new_height=driver.execute_script("return document.body.scrollHeight")
    if new_height==previous_height:
        break
    previous_height=new_height
```

```python
from selenium.webdriver.common.by import By
```

```python
driver.implicitly_wait(100)
```

```python
driver.find_element(By.XPATH,'//html/body/div/div/div/div/div/button[text()="Accept"]').click()
```

```python
content=driver.find_elements(By.TAG_NAME, 'h4')
content[0].text
```

```python
collection_name=[]
for i in range(len(content)):
    collection_name.append(content[i].text)
    
collection_name
```

```python
data=pd.DataFrame(collection_name)
data.rename({0:'Collection list'},inplace=True,axis=1)
data
```

```python
data.to_csv('CollectedData.csv')
```
